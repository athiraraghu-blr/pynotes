```python
time = float(input("Enter the time (0-23): "))
if 5 <= time < 12:
    print("Morning")
elif 12 <= time < 18:
    print("Afternoon")
elif (18 <= time <= 23) or (0 <= time < 5):
    print("Evening")
else:
    print("Invalid hour!")
```

```python
number = int(input("Enter a positive integer: "))
factorial = 1
current = number
while current > 1:
    factorial *= current 
    current -= 1         
print(f"The factorial of {number} is {factorial}")
```
