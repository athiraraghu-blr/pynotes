```python
time = float(input("Enter the time (0-23): "))
if 5 <= time < 12:
    print("Morning")
elif 12 <= time < 18:
    print("Afternoon")
elif (18 <= time <= 23) or (0 <= time < 5):
    print("Evening")
else:
    print("Invalid hour!")
```
